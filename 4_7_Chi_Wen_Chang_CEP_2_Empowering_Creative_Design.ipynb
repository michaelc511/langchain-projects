```python
import gradio as gr
import requests
from PIL import Image
from io import BytesIO
import openai
from google.colab import userdata

print(openai.__version__)

# Retrieve API key securely from Colab's secrets
api_key = userdata.get('OPENAI_API_KEY')

# Initialize the OpenAI client with the API key
client = openai.OpenAI(api_key=api_key)
```

# **Step 2: Define Image Generation Function**

Create function generate_image using OpenAI API for text-to-image process

Before proceeding, please ensure your `OPENAI_API_KEY` is added to Colab's secrets manager. Click the '🔑' icon in the left panel, create a new secret named `OPENAI_API_KEY`, and paste your key there.

```python
import base64

def generate_image(text_prompt):

    try:
        print("Client type:", type(client))
        print("Client:", client)
        print("Prompt:", text_prompt)

        print(f"Debug (Scenario 1): Sending prompt to OpenAI: {text_prompt}") # Debugging line

        response = client.images.generate(
            model="gpt-image-1",
            prompt=text_prompt,
            n=1,
            size="1024x1024"
        )

        print("Response:", response)
        print("=== RESPONSE ===")
        print(response)

        print("=== FIRST DATA ITEM ===")
        print(response.data[0])

        print(f"Debug (Scenario 1): OpenAI API response: {response}") # Debugging line

        if response.data:
            image_data = None
            if response.data[0].url:
                image_url = response.data[0].url
                image_data = requests.get(image_url).content
            elif response.data[0].b64_json:
                image_data = base64.b64decode(response.data[0].b64_json)

            if image_data:
                img = Image.open(BytesIO(image_data))
                return img
            else:
                print("Debug (Scenario 1): No image data found in response.")
                return None
        else:
            print("Debug (Scenario 1): No 'data' in OpenAI response.")
            return None
    except Exception as e:
        print(f"Error (Scenario 1) during image generation: {e}")
        return None
```

# **Step 3: Create Gradio Interface**


*   Set up the Gradio interface with a textbox for prompts and image display
*   Launch the Gradio interface, making it available for user interaction










```python
print(client)
print(type(client))

iface = gr.Interface(
    fn=generate_image,
    inputs=gr.Textbox(label="Enter a text prompt", placeholder="Type a description..."),
    outputs=gr.Image(type="pil"),
    title="Image Generation from Text",
    description="Enter a text description to generate an image."
)

iface.launch(share=True, debug=True)
```

# **Scenario 2:** Logo Designing for a Fictional Company: LogoMaster

# Steps to perform:

1.	Import necessary libraries including Gradio, requests, PIL (Python Imaging Library), BytesIO, and OpenAI
2.	Develop the generate_image function to produce images from text and category using OpenAI API
3.	Set up Gradio interface for easy image generation with user input


```python
import base64 # Added import for base64

# Function to generate image based on text prompt and category
def generate_image(text_prompt, category):


    try:
        prompt_text = ""
        # Determine the prompt based on the selected category
        if category == "Logos":
            prompt_text = "Generate a logo related to " + text_prompt
        elif category == "Company":
            prompt_text = "Generate an image related to the company " + text_prompt

        print(f"Debug: Sending prompt to OpenAI: {prompt_text}") # Debugging line

        # Create an image based on the provided prompt
        response = client.images.generate(
            model="gpt-image-1",
            prompt=prompt_text,
            n=1,
            size="1024x1024"
        )

        print(f"Debug: OpenAI API response: {response}") # Debugging line

        # Check if 'data' key is present in the response
        if response.data:
            image_data = None
            if response.data[0].url:
                image_url = response.data[0].url
                image_data = requests.get(image_url).content
            elif response.data[0].b64_json:
                image_data = base64.b64decode(response.data[0].b64_json)

            if image_data:
                img = Image.open(BytesIO(image_data))
                return img
            else:
                print("Debug: No image data found in response.") # Debugging line
                return None  # Return None if no image data (url or b64_json) is found
        else:
            print("Debug: No 'data' key in OpenAI response.") # Debugging line
            return None  # Return None if 'data' key is not present in the response
    except Exception as e:
        print("Error:", e)
        return None  # Return None in case of any exceptions during image generation

# Create a Gradio interface
iface = gr.Interface(
    fn=generate_image,  # Function to generate the image
    inputs=[
        gr.Textbox(label="Enter a description", placeholder="Type a description..."),
        gr.Radio(choices=["Logos", "Company"], label="Select Category")
    ],
    outputs=gr.Image(type="pil", label="Generated Image"),  # Display the generated image
    title="Image Generation from Text",  # Set interface title
    description="Enter a text description and select a category to generate an image."
)

# Launch the Gradio interface
iface.launch(share=True, debug=True)
```
